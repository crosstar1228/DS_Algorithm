## FIFO인 QUEUE 자료구조로 BFS 구현

```python
def bfs(graph, start_node):
  visited = list() #방문 한 노드
  queue = deque()

  queue.append(start_node) # 시작

  while queue:
    node = queue.popleft()  # 하나씩 나옴
    if node not in visited:
      visited.append(node)
      queue.extend(graph[node])
      print(queue)
  return visited

print(bfs(graph,'A'))
```

## FILO인 stack 자료구조로 DFS 구현
- 따로 모듈을 불러올 것 없이 list의 append, pop활용

```python
def dfs(graph, start_node):
  visited = list()
  stack = list()

  stack.append(start_node)

  while stack:
    node = stack.pop()
    if node not in visited:
      visited.append(node)
      stack.extend(reversed(graph[node]))
      print(stack)
  
  return visited





dfs(graph,'A')
```
